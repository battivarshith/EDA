## Dataset Link: https://www.kaggle.com/datasets/yasserh/titanic-dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

# Dataset Exploration

## Display the first 5 rows of the dataset

```python
print("First 5 rows of the dataset:")
print(df.head())
```

## Check for missing values and summarize the dataset

```python
print("\nMissing values in the dataset:")
print(df.isnull().sum())
```

```python
print("\nSummary of the dataset:")
print(df.info())
```

## Statistics of numerical columns

```python
print("\nBasic statistics of numerical columns:")
print(df.describe())
```

# Visualization 1: Survival Rate by Gender

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=df, palette='pastel')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

# Visualization 2: Survival Rate by Passenger Class

```python
df['Survived'] = df['Survived'].map({0: 'Not Survived', 1: 'Survived'})


plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='muted')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

# Visualization 3: Age Distribution of Survived vs Not Survived

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Survived', y='Age', palette='coolwarm')
plt.title('Age Distribution of Survived vs Not Survived')
plt.xlabel('Survival Status')
plt.ylabel('Age')
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```
